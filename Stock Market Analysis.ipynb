```python
import pandas as pd
```

```python
apple = pd.read_csv('data/stocks/AAPL.csv')
facebook = pd.read_csv('data/stocks/FB.csv')
google = pd.read_csv('data/stocks/GOOGL.csv')
nvidia = pd.read_csv('data/stocks/NVDA.csv')
tesla = pd.read_csv('data/stocks/TSLA.csv')
twitter = pd.read_csv('data/stocks/TWTR.csv')
```

```python
##
dfs = [apple, facebook, google, nvidia, tesla, twitter]
```

```python
#create moving average
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
    
```

```python
apple.head(50)
```

```python
#daily change in gross price
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
